```python
import pandas as pd
import chardet

def read_csv_file(filepath):
  with open(filepath, 'rb') as f:
    result= chardet.detect(f.read())
    encoding = result['encoding']

  try:
    csvpath = pd.read_csv(filepath, encoding=encoding)
    return csvpath
  except UnicodeDecodeError:
    print(f"Error: Unable to read the CSV file with {encoding} encoding.")
    return None

filepath = '/content/drive/MyDrive/이브와ICT공모전/DATASET/경찰청_경찰관서 위치 주소 현황_20220831.csv'

csvpath = read_csv_file(filepath)
csvpath.head()
```

```python
import csv
from geopy.geocoders import Nominatim

def geocode_address(address):
    geolocator = Nominatim(user_agent="address_geocoder")
    location = geolocator.geocode(address)
    if location is not None:
        return location.latitude, location.longitude, location.address
    else:
        return None, None, None
    #location 객체가 None이 아닌 경우에만 위도, 경도, 전체 주소 정보를 반환

def geocode_csv(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as csv_in, \
         open(output_file, 'w', newline='', encoding='utf-8') as csv_out:
        reader = csv.reader(csv_in)
        writer = csv.writer(csv_out)
        header = next(reader)
        writer.writerow(header + ['Latitude', 'Longitude', 'Full Address'])

        for row in reader:
            address = row[7]  #인덱스 0으로 시작하는 기준으로 7번째 열의 정보를 위도경도로 변환하여 8번째 열부터 새롭게 저장하여 추가
            latitude, longitude, full_address = geocode_address(address)
            writer.writerow(row + [latitude, longitude, full_address])

if __name__ == '__main__':
    input_csv_file = '/content/drive/MyDrive/이브와ICT공모전/DATASET/경찰청_경찰관서 위치 주소 현황.csv'   # Replace with the name of your input CSV file
    output_csv_file = '/content/drive/MyDrive/이브와ICT공모전/DATASET/경찰청_경찰관서 위치 주소 현황_지오코딩.csv' # Replace with the desired name for the output CSV file
    geocode_csv(input_csv_file, output_csv_file)
```
